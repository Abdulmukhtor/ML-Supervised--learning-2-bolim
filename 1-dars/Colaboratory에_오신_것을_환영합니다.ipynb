```
x=df.drop(columns=['personal_loan'])
```

```
y=df['personal_loan']
```

```
df.head()
```

Datalarni train, test, splitga ajratish

```
x_train,x_temp,y_train,y_temp=train_test_split(x,y,test_size=0.2,random_state=42)
x_test,x_val,y_test,y_val=train_test_split(x_temp,y_temp,test_size=0.5,random_state=42)
```

Modelni tanlash va o'rgatish

```
model=LogisticRegression() # modelni tanlash
Logistic_model=model.fit(x_train,y_train) # modelni x va y train orqali o'rgatish
y_pred = Logistic_model.predict(x_test) # modelni train qilib x_test oxirgi natijani olish
acc=accuracy_score(y_test,y_pred)
```

```
print(f"Model aniqligi:{int(acc*100)}%")
```
